# PharmaPulse Commercial Analytics

## Commercial Analytics for Pharmaceutical Product Demand

### Project Overview

This project analyzes historical pharmaceutical sales data to identify commercial trends, demand patterns, and high-performing therapeutic drug categories.

The objective is to transform raw sales data into business insights that can support inventory planning, forecasting, and strategic commercial decision-making.

---

## Business Objectives

The analysis aims to answer the following questions:

- Which drug category contributes the highest sales?
- Which months experience peak pharmaceutical demand?
- Are there seasonal sales trends?
- Which weekdays generate the highest sales?
- How has sales performance changed over the years?
- Which commercial insights can improve inventory management?

---
